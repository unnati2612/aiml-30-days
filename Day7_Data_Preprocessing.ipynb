```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Student_ID": [101, 102, 103, 104, 105, 106],
    "Marks": [85, np.nan, 78, 92, 500, 88],
    "Interest": [7, 8, np.nan, 9, 6, 7],
    "Branch": ["CS", "IT", "CS", "IT", "CS", "CS"],
    "Selected": ["Yes", "Yes", "No", "Yes", "No", np.nan]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset = ["Selected"])
df
```

```python
df["Marks"].fillna(df["Marks"].median(), inplace = True)
```

```python
df
```

```python
df["Interest"].fillna(df["Interest"].median(), inplace=True)
```

```python
df
```

```python
df = df[df["Marks"]<=100]
```

```python
df.drop("Student_ID" , axis=1, inplace = True)
df
```

```python
df["Selected"].unique()
```

```python
df["Selected"] = df["Selected"].map({"Yes":1,"No":0})
df
```

```python
df = pd.get_dummies(df,columns = ["Branch"] , drop_first= True )
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df.info()
```

```python
df[["Marks","Interest"]] = scaler.fit_transform(df[["Marks","Interest"]])
```

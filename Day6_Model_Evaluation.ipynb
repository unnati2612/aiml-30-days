# Day 6 – Model Evaluation & Metrics

## Objective
Learn how to evaluate machine learning models properly instead of relying only on accuracy.

## Concepts Covered
- Confusion Matrix
- Accuracy, Precision, Recall
- False Positives vs False Negatives
- Overfitting and Underfitting
- Bias–Variance Trade-off
- Metric selection based on business risk

## Key Insights
- Accuracy is misleading for imbalanced datasets.
- Precision is important when false positives are costly.
- Recall is critical when missing positive cases is dangerous.
- Model evaluation should align with real-world business impact.

## Example Scenarios
- Fraud detection → prioritize Recall
- Spam detection → prioritize Precision
- Balanced datasets → Accuracy is acceptable

## Conclusion
This work helped me develop an evaluation mindset required for real-world machine learning systems.

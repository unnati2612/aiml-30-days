```python
def assign_grades(Marks):
  if Marks>= 90:
    return "A"
  elif Marks>= 80:
    return "B"
  else:
    return "C"
```

```python
df["Grade"] = df["Marks"].apply(assign_grades)
df
```

# Day 1 â€“ Pandas Basics (Explanation & Learning)

## What is Pandas?
Pandas is a Python library used to work with structured data.
It helps us create tables (DataFrames) similar to Excel sheets and perform data analysis.

---

## Importing Pandas
`import pandas as pd`

- This line imports the pandas library
- `pd` is a short name used for convenience

---

## Creating Data using Dictionary
We created a Python dictionary where:
- Keys are column names
- Values are lists of data for each column

Example:
- Name â†’ student names
- Age â†’ student ages
- Marks â†’ student marks

---

## Creating a DataFrame
`df = pd.DataFrame(data)`

- Converts the dictionary into a DataFrame
- A DataFrame is a 2D table with rows and columns

---

## Viewing the Data
`df`

- Displays the complete table in the notebook

---

## df.head()
`df.head()`

- Shows the first 5 rows of the dataset
- Used to quickly check if data loaded correctly

---

## df.describe()
`df.describe()`

- Gives statistical summary of numerical columns
- Includes:
  - count
  - mean (average)
  - min and max values
  - percentile values (25%, 50%, 75%)

---

## df.info()
`df.info()`

- Shows:
  - Number of rows and columns
  - Column names
  - Data types (int, float, object)
  - Whether there are missing values or not

---

## Creating a New Column (Passed)
`df["Passed"] = df["Marks"] >= 80`

- Creates a new column named "Passed"
- If Marks â‰¥ 80 â†’ True
- Else â†’ False
- This is an example of applying conditions on data

---

## Filtering Data
`df[df["Passed"] == True]`

- Shows only those rows where students have passed
- Used to filter data based on conditions

---

## Mean of Marks
`df["Marks"].mean()`

- Calculates the average marks of all students

---

## Sorting Data
`df.sort_values(by="Marks", ascending=False)`

- Sorts the data based on Marks
- Highest marks appear first

---

## Grade Column Logic
We created a Grade column using conditions:
- Marks â‰¥ 90 â†’ Grade A
- Marks â‰¥ 80 â†’ Grade B
- Else â†’ Grade C

This was done using a function and applying it to the Marks column.

---

## Key Learning from Day 1
- How to create and view a DataFrame
- How to analyze data using Pandas
- How to add new columns using conditions
- How to filter and sort data
